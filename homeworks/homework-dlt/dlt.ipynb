```python
import dlt
from dlt.sources.helpers.rest_client import RESTClient
from dlt.sources.helpers.rest_client.paginators import PageNumberPaginator


@dlt.resource
def data_downloader():

    client = RESTClient(base_url='https://us-central1-dlthub-analytics.cloudfunctions.net/data_engineering_zoomcamp_api',
    paginator=PageNumberPaginator(base_page=1,total_path=None,stop_after_empty_page=True),
    )
    for page in client.paginate():
        yield page



pipeline = dlt.pipeline(
    pipeline_name="ny_taxi_pipeline",
    destination="duckdb",
    dataset_name="ny_taxi_data"
)
```

```python
load_info = pipeline.run(data_downloader)
print(load_info)
```

```python
import duckdb

# A database '<pipeline_name>.duckdb' was created in working directory so just connect to it

# Connect to the DuckDB database
conn = duckdb.connect("ny_taxi_pipeline.duckdb")


tables = conn.sql("SHOW TABLES").fetchdf()
print("Tables in the database:", tables)

# Describe the dataset
conn.sql("DESCRIBE").df()
```
