```python
import pandas as pd 
from sqlalchemy import create_engine
from time import time
```

```python
df = pd.DataFrame(pd.read_csv("yellow_tripdata_2021-01.csv"))
df
```

```python
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
df
```

```python
engin = create_engine('postgresql://root:root@localhost:5432/ny_taxi_data')
engin.connect()
```

```python
sql_table = pd.io.sql.get_schema(df,name='yellow_taxi_data',con=engin)
```

```python
print(sql_table)
```

```python
df_iter  = pd.read_csv('yellow_tripdata_2021-01.csv',iterator=True,chunksize=100000)
```

```python
df_iter
```

```python
df = next(df_iter)
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
```

```python
len(df)
```

```python
df.info()
```

```python
df.head(n=0).to_sql(name='yellow_taxi',con=engin,if_exists='replace')
```

```python
df.to_sql(name='yellow_taxi',con=engin,if_exists='append')
```

```python
while True:
    try :
        begin = time()
        df = next(df_iter)
        df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
        df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime'])
        df.to_sql(name='yellow_taxi',con=engin,if_exists='append')
        end = time()
    
        print(f'tHE NEW CHUNCK WAS ADDED TO THE DATABASE IN : {end-begin} SECONDS')
        
    except StopIteration:

        print('NO MORE DATA TO APPEND')
        break
        
    
```

```python
query = 'SELECT * FROM yellow_taxi LIMIT 10'
pd.read_sql(sql=query,con=engin)
```
